```python
import requests
import mwparserfromhell
import json
import time

# Next time:: change links to be usefull change to the https://en.wikipedia.org/index.php/{title}&action=raw format

headers = {
    'User-Agent': 'MyCoasterProject/1.0 (your_email@example.com)'
}

links = {
    "https://als.wikipedia.org/wiki/G'sengte_Sau",
    "https://de.wikipedia.org/wiki/G’sengte_Sau",
    "https://nl.wikipedia.org/wiki/G'sengte_Sau",
    "https://commons.wikipedia.org/wiki/Category:G'sengte_Sau",
    "https://fr.wikipedia.org/wiki/G'sengte_Sau"
}

r = requests.get(
    f"{links[0]}"
    'https://en.wikipedia.org/w/index.php?title=Fury_325&action=raw',
    headers=headers
)

#print(r.text[:200])

wikicode = mwparserfromhell.parse(r.text)
templates = wikicode.filter_templates()

rinfobox = None
for template in templates:
    if str(template.name).strip() == "Infobox roller coaster":
        rinfobox = template

if rinfobox:
    print(f"Found template: {rinfobox.name}")


data = {}
for param in rinfobox.params:
    name = str(param.name).strip()
    value = str(param.value).strip()
    data[name] = value

print(data)
    


```
